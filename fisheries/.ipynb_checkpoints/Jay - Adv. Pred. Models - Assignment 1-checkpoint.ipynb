###  2.1 Write a function to compute the mean squared error between a prediction and ground truth assuming both are numpy arrays (see python module numpy).

```python
def mse (pred, y):
    err = y-pred
    mse = (np.multiply(err, err))/y.shape[0]
    return np.matrix(mse)
```

### 2.4 Assuming that you do not know the model, use backpropagation and gradient descent updates to find the model parameters (choose an appropriate learning rate). The loss function will be the mean squared error.

```python
def grad_descent (X, y, reg = 1e-4, step_size = 1e-1, \
                  stop = 1e-5, iter = 1000, mode='S'):
    
    # Linear model, No hidden layers
    D = 1
    K = 1 # There is only one output node

    # Initialize parameters randomly
    W = 0.05 * np.random.randn(D,K)
    b = np.zeros((1,K))

    batch_size = X.shape[0]
    mse_list = []
    #Perform batch SGD using backprop
    for i in range(iter):

        # evaluate class scores, [N x K]
        scores = np.dot(X, W) + b 

        # Compute the loss
        data_loss = np.sum(mse(y, scores))
        mse_list.append(data_loss)

        reg_loss = 0.5*reg*np.sum(W*W)
        loss = data_loss + reg_loss
        if loss < stop: break

            
        if (mode.upper() == 'V') and \
        ((i == 0) or (i == iter-1) or (i % 100 == 0)):
            print ("iteration {}: loss = {}".format(i, loss))
            
        # compute the gradient on scores
        dscores = -(y-scores)/batch_size 

        # backpropagate the gradient to the parameters (W,b)
        dW = np.dot(X.T, dscores)
        db = np.sum(dscores, axis=0)

        dW += reg*W # regularization gradient


        # perform a parameter update
        W += -step_size * dW
        b += -step_size * db
    
    if (mode.upper() == 'V'):
        print ('Abs Diff = ', np.sum(abs(y - scores)))
        print ('W = ', W)
        print ('b =', b)

    return W, b, mse_list
```

### 3.1 Write a function to compute the logistic loss given the parameters (W, b) of a linear model (as numpy arrays) and an example (x, y).

```python
def logistic_loss (W, b, x, y):
    '''
    Assumes y is a scalar and has the right class
    encoded as integer ranging between 0 and n-1
    '''
    
    scores = np.dot(x, W) + b
    
    # compute the class probabilities
    exp_pred = np.exp(scores)
    probs = exp_pred / np.sum(exp_pred, axis=1, keepdims=True)
  
    # logistic loss is the negative log of  
    # prob of the predicted value of the correct class
    log_loss = -np.log(probs[0,y])
    
    return log_loss
```

### 3.2 Write a function to compute the multiclass SVM loss given the parameters (W, b) of a linear model (as numpy arrays) and an example (x, y).

```python
def svm_loss (W, b, x, y, c=0):
    '''
    Assumes y is a scalar and has the right class
    encoded as integer ranging between 0 and n-1
    '''
    scores = np.dot(x, W) + b
    correct_class_score = scores[0, y]

    svm_loss = np.max(0, (scores - correct_class_score)  + c)
    #Correct for the extra c (of the right class) that got added in above step 
    svm_loss -= c
    
    return svm_loss
```


### 3.3 Add an L1 regularization and an L2 regularization to the loss function.

```python

def loss (W, b, X, y, c=0, alpha = 0.5, loss_type='LOG'):
    '''
    Assumes y is a scalar and has the right class
    encoded as integer ranging between 0 and n-1
    '''
    scores = np.dot(X, W) + b
    
    if loss_type.upper() == 'SVM':
        # SVM Loss
        correct_class_score = scores[0, y]
        loss = np.max(0, (scores - correct_class_score)  + c)
        #Correct for the extra c (of the right class) that got added in above step 
        loss -= c
    elif loss_type.upper() == 'LOG':
        # Logistic loss - compute the class probabilities
        exp_pred = np.exp(scores)
        probs = exp_pred / np.sum(exp_pred, axis=1, keepdims=True)

        # logistic loss is the negative log of  
        # prob of the predicted value of the correct class
        loss = -np.log(probs[0,y])

    
    # Apply L1 and L2 Regularization
    L1_reg = np.sum(np.abs(W)) + np.sum(np.abs(b))
    L2_reg = np.sum(np.square(W)) + np.sum(np.square(b))
    elast_net = alpha * L1_reg + (1 - alpha)*L2_reg

    # Add elastic net to loss
    loss += elast_net

    
    return loss
```

### 4.3 Build a linear classifier assuming the multiclass logistic loss and an l2 regularization for the weights only. Report the prediction accuracy on the training data and the test data.

```python
class clf_linear(object):

    def __init__(self, b_size = 1, n_iter = 500, step_size = 1e-0, \
                 reg = 1e-4, stop = 1e-4, mode = 'S', seed=0):
    
        self.name = 'Linear Clasfr'
        self.b_size = b_size
        self.n_iter = n_iter
        self.step_size = step_size
        self.reg = reg
        self.seed = seed
        self.stop = stop
        self.mode = mode

    def fit(self, X_trg, y_trg, reg = 0, step_size = 0, n_iter = 0):
        
        np.random.seed(self.seed)
        
        stop = self.stop
        mode = self.mode
        b_size = self.b_size
        
        if (n_iter == 0):
            n_iter = self.n_iter
        else:
            self.n_iter = n_iter
            
        if (reg == 0):
            reg = self.reg
        else:
            self.reg = reg
        
        if (step_size == 0):
            step_size = self.step_size
        else:
            self.step_size = step_size
        
        D = X_trg.shape[1] #Number of features
        K = max(y_trg)+1 #Number of classes assuming class index starts from 0

        # Start with an initialize parameters randomly
        self.W = 0.01 * np.random.randn(D,K)
        self.b = np.zeros((1,K))

        #Perform batch SGD using backprop
        batch_size = X_trg.shape[0]

        # gradient descent loop
        for i in range(n_iter):

            # evaluate class scores, [N x K]
            scores = np.dot(X_trg, self.W) + self.b 

            # Logistic loss - compute the class probabilities
            exp_scores = np.exp(scores)
            probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]

            # compute the loss: average cross-entropy loss and regularization
            corect_logprobs = -np.log(probs[range(batch_size),y_trg])
            data_loss = np.sum(corect_logprobs)/batch_size

            # L2 Regularization on W
            reg_loss = 0.5*reg*np.sum(self.W*self.W)
            loss = data_loss + reg_loss

            if (self.mode.upper() != 'S') and \
            ((i == 0) or (i == n_iter-1) or (i % 100 == 0)):
                print ("iteration {}: loss = {}".format(i, loss))

            if loss < self.stop: break

            # compute the gradient on scores
            dscores = probs
            dscores[range(batch_size),y_trg] -= 1
            dscores /= batch_size

            # backpropate the gradient to the parameters (W,b)
            dW = np.dot(X_trg.T, dscores)
            db = np.sum(dscores, axis=0, keepdims=True)

            dW += reg*self.W # regularization gradient

            # perform a parameter update
            self.W += -step_size * dW
            self.b += -step_size * db

    def score(self, X, y):
        # Run on test data
        scores = np.dot(X, self.W) + self.b
        predicted_class = np.argmax(scores, axis=1)
        acc = (np.mean(predicted_class == y))
        return acc
```

### 4.4 Introduce a cross validation scheme and justify your choice. What is the validation accuracy compare to the test accuracy.

#### We use Cross validation  because - 
##### 1) It allows the use of training data for both training and validation of the model.  2) We can ascertain the stabilty of model built with a different set of inputs in each fold.  3) It could also be used to tune model parameters while avoiding over-fitting.
#### We use K-fold cross validation here as  - It allows easy split of boot strapped training samples
######
#### We will find the best split that has a mean validation accuracy that closely matches the accuracy of test data. This should allow us to use the best matching split to tune hyper parameters and the results could be extrapolated to the test set

```python
# Cross validate model using multiple splits of training data
def cv_model (clf, X_trg, y_trg, X_tst, y_tst, \
                     tst_size=0.2, splits_max = 5):
    
    # First get the test data accuracy while trained on full trg data
    clf.fit(X_trg, y_trg)
    acc_tst = clf.score(X_tst, y_tst)
    
    # Hold the validation accuracies for each split
    mean_acc = np.zeros(shape = (splits_max-1,2), dtype=float)
        
    # Iterate over multiple splits to find the best split number
    splits = range(2, splits_max+1, 1)
    for split in splits:
        
        #Use K-fold Cross Validation
        kf = KFold(n_splits=split)
        
        # Train using each fold
        loop_indx=0
        loop_acc_split = np.zeros(split, dtype=float)
        loop_acc_tst = np.zeros(split, dtype=float)
        for tr_indx, tst_indx in kf.split(X_train):
            # Fit model with trg data from split
            clf.fit(X_trg[tr_indx], y_trg[tr_indx])
            
            # Score model with test data from split
            loop_acc_split[loop_indx] = clf.score(X_trg[tst_indx], y_trg[tst_indx])
            
            # Score model with original test data
            loop_acc_tst[loop_indx] = clf.score(X_tst, y_tst)
            loop_indx += 1
        mean_acc[split-2,0] = loop_acc_split.mean()
        mean_acc[split-2,1] = loop_acc_tst.mean()
        
        print ("Mean Validation accuracy across %d splits = %f" \
               %(split, mean_acc[split-2,0]))
        print ("Mean Test accuracy across %d splits = %f" \
               %(split, mean_acc[split-2,1]))
    
#     print ('Accuracy on test data trained with full training data= ', acc_tst)
    
    print ('')
    print ('Best Number of splits that closeley matches test data accuracy = %d' \
    %(splits[np.argmin(np.abs(mean_acc[:, 0] - acc_tst))]))
    print ('Test Data Accuracy = %.4f' %(acc_tst))
    print ('Best Match Validation Accuracy = %.4f' \
           %(mean_acc[np.argmin(np.abs(mean_acc[:,0] - acc_tst))])[0])
```

##### Model is stable becuase - 
##### 1) Mean Validation accuracies across multiple splits are similar.
##### 2) Mean test accuracies for models built on each split are similar to each other, similar to validation accuracies as well as similar to the full test accuracy


### 4.5 What is the sensitivity of the model’s performance to different learning rates and the number of gradient descent iterations. Describe via suitable plots.

```python
# Sensitivity to Learning Rates

def plot_lrn_rate_senstvty (clf, X_trg, y_trg, X_tst, y_tst, eta_min, eta_max, \
                            eta_step, iter1 = 100, iter2 = 200, reg= 1e-3):
    # Sensitivity to learning rates
    etas = np.arange(eta_min, eta_max+eta_step, eta_step)
    # Loop over different learning rates
    acc1 = np.zeros(etas.size)
    acc2 = np.zeros(etas.size)
    loop_indx = 0
    for eta in etas:
        
        # Model1 with iter1 iterations
        clf.fit(X_trg, y_trg, reg = reg, step_size = eta, n_iter=iter1)
        acc1[loop_indx] = clf.score(X_tst, y_tst)
        
        # Model2 with iter2 iterations
        clf.fit(X_trg, y_trg, reg = reg, step_size = eta, n_iter=iter2)
        acc2[loop_indx] = clf.score(X_tst, y_tst)
        
        loop_indx += 1

    #Plot
    title = 'Test Accuracy Vs Learning Rate (' + clf.name + ')'
    plt.title(title)
    plt.ylabel('Test Accuracy')
    plt.xlabel('Learning Rate')

    plt.plot(etas, acc1, 'b-')
    plt.plot(etas, acc2, 'r-')
    
#     xpos = etas[int(0.7*len(etas))]
    xpos = min(etas) + 0.75*(max(etas) - min(etas))
    y_min = min(min(acc1), min(acc2))
    y_max = max(max(acc1), max(acc2))
    ypos1 = y_min + 0.92*(y_max - y_min)
    ypos2 = y_min + 0.85*(y_max - y_min)
    
    plt.annotate('Iterations = {}'.format(iter1), color='b', \
                 xy=(xpos, ypos1), xytext=(xpos, ypos1)) 
    
    plt.annotate('Iterations = {}'.format(iter2), color='r', \
                 xy=(xpos, ypos1), xytext=(xpos, ypos2))    
    plt.show()
    
    print ("Iterations %d: Max Test accuracy %.4f occurs with learning rate %.5f"\
           %(iter1, np.max(acc1), etas[np.argmax(acc1)]))
    print ("Iterations %d: Min Test accuracy %.4f occurs with learning rate %.5f"\
           %(iter1, np.min(acc1), etas[np.argmin(acc1)]))
    
    print ("Iterations %d: Max Test accuracy %.4f occurs with learning rate %.5f"\
           %(iter2, np.max(acc2), etas[np.argmax(acc2)]))
    print ("Iterations %d: Min Test accuracy %.4f occurs with learning rate %.5f"\
           %(iter2, np.min(acc2), etas[np.argmin(acc2)]))
```

```python
# Plot Sensitivity to number of iterations for different learn rates
def plot_iter_senstvty (clf, X_trg, y_trg, X_tst, y_tst, iter_min, \
                        iter_max, iter_step, eta1 = 1e-5, eta2 = 5e-1, \
                        reg=1e-3):
    
    iters = np.arange(iter_min, iter_max+iter_step, iter_step)
    
    acc1 = np.zeros(iters.size)
    acc2 = np.zeros(iters.size)
    loop_indx = 0
    for itr in iters:
        # Model1 with iter1 iterations
        clf.fit(X_trg, y_trg, reg = reg, step_size = eta1, n_iter=itr)
        acc1[loop_indx] = clf.score(X_tst, y_tst)
        
        # Model2 with iter2 iterations
        clf.fit(X_trg, y_trg, reg = reg, step_size = eta2, n_iter=itr)
        acc2[loop_indx] = clf.score(X_tst, y_tst)
        
        loop_indx += 1

    #Plot
    title = 'Test Accuracy Vs Iterations ('+ clf.name + ')'
    plt.title(title)
    plt.ylabel('Test Accuracy')
    plt.xlabel('Iterations')

    plt.plot(iters, acc1, 'b-')
    plt.plot(iters, acc2, 'r-')
    
#     xpos = etas[int(0.7*len(etas))]
    xpos = min(iters) + 0.7*(max(iters) - min(iters))
    y_min = min(min(acc1), min(acc2))
    y_max = max(max(acc1), max(acc2))
    ypos1 = y_min + 0.92*(y_max - y_min)
    ypos2 = y_min + 0.85*(y_max - y_min)
    
    
    plt.annotate('Learning Rate = {}'.format(eta1), color='b', \
                 xy=(xpos, ypos1), xytext=(xpos, ypos1))
    
    plt.annotate('Learning Rate = {}'.format(eta2), color='r', \
                 xy=(xpos, ypos1), xytext=(xpos, ypos2)) 
    
    plt.show()
    
    print ("Learn Rate %.4f: Max Test accuracy %.4f occurs at iteration %d"\
           %(eta1, np.max(acc1), iters[np.argmax(acc1)]))
    print ("Learn Rate %.4f: Min Test accuracy %.4f occurs at iteration %d"\
           %(eta1, np.min(acc1), iters[np.argmin(acc1)]))
    
    print ("Learn Rate %.4f: Max Test accuracy %.4f occurs at iteration %d"\
           %(eta2, np.max(acc2), iters[np.argmax(acc2)]))
    print ("Learn Rate %.4f: Min Test accuracy %.4f occurs at iteration %d"\
           %(eta2, np.min(acc2), iters[np.argmin(acc2)]))
```

```python
# Sensitivity to regularization
def get_best_reg (clf, X_trg, y_trg, X_tst, y_tst, reg_min = 1e-4, reg_max = 1e0, \
              reg_step = 1e-4, plot=True):

    regs = np.arange(reg_min, reg_max+reg_step, reg_step)
    # Loop over different Reg rates
    loop_indx = 0
    acc = np.zeros(regs.size)
    for reg in regs:
        
        # Fit the model
        clf.fit(X_train, y_train, reg=reg)

        acc[loop_indx] = clf.score(X_test, y_test)
        loop_indx += 1

    if(plot):
        #Plot
        plt.title('Test Accuracy Vs Regularization')
        plt.ylabel('Test Accuracy')
        plt.xlabel('Regularization')

        plt.plot(regs, acc, 'g-')
        plt.show()
    
    # Run with zero reg and get accuracy
    clf.fit(X_train, y_train, reg=0)
    zero_acc = clf.score (X_tst, y_tst)
    
    best_reg = regs[np.argmax(acc)]
    best_acc = np.max(acc)
    
    print ('Best Reg (at learn rate: {}, Iter: {}) = {}'\
           .format(clf.step_size, clf.n_iter, best_reg) )
    print ('Accuracy with best reg = %.4f'%(best_acc))
    print ('Accuracy with ZERO reg = %.4f'%(zero_acc))
```

### 5.1 Repeat Question 4 when the model is a 2-layer feedforward neural network (i.e.,1 hidden layer with f (x, W 1 , b 1 , W 2 , b 2 ) = W 2 max(0, W 1 x + b 1 ) + b 2. Use hand-computed backpropagation equations.

##### 5.1.3 Build an FFN classifier assuming the multiclass logistic loss and an L2 regularization for the weights only. Report the prediction accuracy on the training data and the test data.

```python
# 2Layer FFN using Relu
class clf_relu(object):

    def __init__(self, h = 100, n_iter = 5000, step_size = 1e-0, reg=1e-4,\
                 b_size=2, stop = 1e-4, mode = 'S', seed=0):
        self.name = 'ReLU'
        self.seed = seed
        self.h = h
        self.n_iter = n_iter
        self.step_size = step_size
        self.reg = reg
        self.b_size = b_size
        self.stop = stop
        self.mode = mode

    def fit(self, X_trg, y_trg, reg = 0, step_size = 0, n_iter = 0):
        
        np.random.seed(self.seed)
        
        h = self.h
        stop = self.stop
        mode = self.mode
        b_size = self.b_size
        
        if (n_iter == 0):
            n_iter = self.n_iter
        else:
            self.n_iter = n_iter
        
        if (step_size == 0):
            step_size = self.step_size
        else:
            self.step_size = step_size
        
        if (reg == 0):
            reg = self.reg
        else:
            self.reg = reg
        
        #Define some local varaibles
        D = X_trg.shape[1] #Number of features
        K = np.unique(y_trg).size #Number of classes assuming class index starts from 0

        # Start with an initial set of parameters randomly
        self.W1 = 0.01 * np.random.randn(D,h)
        self.b1 = np.zeros((1,h))
        self.W2 = 0.01 * np.random.randn(h,K)
        self.b2 = np.zeros((1,K))

        # Set batch size
        batch_size = int(X_trg.shape[0]/b_size)
        max_row = X_trg.shape[0]  #Tot rows in training data
        batch_count=0
        for row in range(0, max_row, batch_size):
            #Compute batch to process
            tmp = row + batch_size
            if tmp > max_row:
                batch_max = max_row
                batch_size = batch_max - row
            else:
                batch_max = tmp

            # Get the batch of data points
            X_batch = X_trg[row:batch_max]
            y_batch = y_trg[row:batch_max]

            batch_count += 1
            if mode.upper()=='V': 
                print("Batch Count:", batch_count)

            # gradient descent loop
            for i in range(n_iter):

                # evaluate class scores, [N x K]
                hl_out = np.maximum(0, np.dot(X_batch, self.W1) + self.b1) # note, ReLU activation
                scores = np.dot(hl_out, self.W2) + self.b2

                # compute the class probabilities
                exp_scores = np.exp(scores)
                probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]

                # compute the loss: average cross-entropy loss and regularization
                corect_logprobs = -np.log(probs[range(batch_size),y_batch])
                data_loss = np.sum(corect_logprobs)/batch_size
                reg_loss = 0.5*reg*np.sum(self.W1*self.W1) + 0.5*reg*np.sum(self.W2*self.W2)
                loss = data_loss + reg_loss
                if (mode.upper()=='V') and ((i % (n_iter/10) == 0) or (i % (n_iter-1) == 0)):
                    print ("iteration {}: loss {}".format(i, loss))

                if loss < stop : break

                # compute the gradient on scores
                dscores = probs
                dscores[range(batch_size),y_batch] -= 1
                dscores /= batch_size

                # backpropate the gradient to the parameters
                # first backprop into parameters W2 and b2
                dW2 = np.dot(hl_out.T, dscores)
                db2 = np.sum(dscores, axis=0, keepdims=True)

                # next backprop into hidden layer
                dhidden = np.dot(dscores, self.W2.T)
                # backprop the ReLU non-linearity
                dhidden[hl_out <= 0] = 0

                # finally into W,b
                dW1 = np.dot(X_batch.T, dhidden)
                db1 = np.sum(dhidden, axis=0, keepdims=True)

                # add regularization gradient contribution
                dW2 += reg * self.W2
                dW1 += reg * self.W1

                # perform a parameter update
                self.W1 += -step_size * dW1
                self.b1 += -step_size * db1
                self.W2 += -step_size * dW2
                self.b2 += -step_size * db2

    def score(self, X_tst, y_tst):
        # Run on test data
        hl_out = np.maximum(0, np.dot(X_tst, self.W1) + self.b1)
        scores = np.dot(hl_out, self.W2) + self.b2
    
        predicted_class = np.argmax(scores, axis=1)
        acc = (np.mean(predicted_class == y_tst))
    
        return acc
```

### 5.2 Repeat item 1 when the non-linearity is changed to leaky ReLU (f (x) = max(0.01∗x, x)).

#### 5.2.3 Build an FFN classifier with Leaky ReLU as the activation function.

```python
# 2Layer FFN using Leaky Relu
class clf_leaky_relu(object):

    def __init__(self, h = 100, n_iter = 5000, step_size = 1e-0, reg=1e-4,\
                 b_size=2, stop = 1e-4, mode = 'S', seed=0):
        self.name = 'Leaky ReLU'
        self.seed = seed
        self.h = h
        self.n_iter = n_iter
        self.step_size = step_size
        self.reg = reg
        self.b_size = b_size
        self.stop = stop
        self.mode = mode

    def fit(self, X_trg, y_trg, reg = 0, step_size = 0, n_iter = 0):
        
        np.random.seed(self.seed)
        
        h = self.h
        stop = self.stop
        mode = self.mode
        b_size = self.b_size
        
        if (n_iter == 0):
            n_iter = self.n_iter
        else:
            self.n_iter = n_iter
        
        if (step_size == 0):
            step_size = self.step_size
        else:
            self.step_size = step_size
        
        if (reg == 0):
            reg = self.reg
        else:
            self.reg = reg
        
        #Define some local varaibles
        D = X_trg.shape[1] #Number of features
        K = np.unique(y_trg).size #Number of classes assuming class index starts from 0

        # Start with an initial set of parameters randomly
        self.W1 = 0.01 * np.random.randn(D,h)
        self.b1 = np.zeros((1,h))
        self.W2 = 0.01 * np.random.randn(h,K)
        self.b2 = np.zeros((1,K))

        # Set batch size
        batch_size = int(X_trg.shape[0]/b_size)
        max_row = X_trg.shape[0]  #Tot rows in training data
        batch_count=0
        for row in range(0, max_row, batch_size):
            #Compute batch to process
            tmp = row + batch_size
            if tmp > max_row:
                batch_max = max_row
                batch_size = batch_max - row
            else:
                batch_max = tmp

            # Get the batch of data points
            X_batch = X_trg[row:batch_max]
            y_batch = y_trg[row:batch_max]

            batch_count += 1
            if mode.upper()=='V': 
                print("Batch Count:", batch_count)

            # gradient descent loop
            for i in range(n_iter):

                # evaluate class scores, [N x K]
                tmp = np.dot(X_batch, self.W1) + self.b1
                hidden_layer_out = np.maximum(0.01*tmp, tmp) # note, Leaky ReLU activation
                scores = np.dot(hidden_layer_out, self.W2) + self.b2

                # compute the class probabilities
                exp_scores = np.exp(scores)
                probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]

                # compute the loss: average cross-entropy loss and regularization
                corect_logprobs = -np.log(probs[range(batch_size),y_batch])
                data_loss = np.sum(corect_logprobs)/batch_size
                reg_loss = 0.5*reg*np.sum(self.W1*self.W1) \
                + 0.5*reg*np.sum(self.W2*self.W2)
                loss = data_loss + reg_loss
                if (mode.upper()=='V') and ((i % (n_iter/10) == 0) or (i % (n_iter-1) == 0)):
                    print ("iteration {}: loss {}".format(i, loss))

                # compute the gradient on scores
                dscores = probs
                dscores[range(batch_size),y_batch] -= 1
                dscores /= batch_size

                # backpropate the gradient to the parameters
                # first backprop into parameters W2 and b2
                dW2 = np.dot(hidden_layer_out.T, dscores)
                db2 = np.sum(dscores, axis=0, keepdims=True)

                # next backprop into hidden layer
                dhidden = np.dot(dscores, self.W2.T)
                # backprop the Leaky ReLU non-linearity
                tmp = np.dot(X_batch, self.W1) + self.b1
                dhidden[hidden_layer_out > tmp] *= 0.01

                # finally into W,b
                dW1 = np.dot(X_batch.T, dhidden)
                db1 = np.sum(dhidden, axis=0, keepdims=True)

                # add regularization gradient contribution
                dW2 += reg * self.W2
                dW1 += reg * self.W1

                # perform a parameter update
                self.W1 += -step_size * dW1
                self.b1 += -step_size * db1
                self.W2 += -step_size * dW2
                self.b2 += -step_size * db2

    def score(self, X_tst, y_tst):
        # Run on test data        
        tmp = np.dot(X_tst, self.W1) + self.b1
        hl_out = np.maximum(0.01*tmp, tmp) # note, Leaky ReLU activation
        scores = np.dot(hl_out, self.W2) + self.b2

        predicted_class = np.argmax(scores, axis=1)
        acc = (np.mean(predicted_class == y_tst))
    
        return acc
```

### 5.3 Repeat Question 4 when the model is f (x, W 1 , b 1 , W 2 , b 2 , W 3 , b 3 ) = W 3 max(W 2 x + b 2 , W 1 x + b 1 ) + b 3 (this is called the maxout non-linearity).

##### 5.3.3 Build a linear classifier assuming the multiclass logistic loss and an l2 regularization for the weights only.

```python
# 2Layer FFN using maxout activation
class clf_maxout(object):

    def __init__(self, h = 100, n_iter = 5000, step_size = 1e-0, reg=1e-4, \
                 b_size=2, stop = 1e-4, mode = 'S', seed=0):
        self.name = 'Maxout'
        self.seed = seed
        self.h = h
        self.n_iter = n_iter
        self.step_size = step_size
        self.reg = reg
        self.b_size = b_size
        self.stop = stop
        self.mode = mode

    def fit(self, X_trg, y_trg, reg = 0, step_size = 0, n_iter = 0):
        
        np.random.seed(self.seed)
        
        h = self.h
        stop = self.stop
        mode = self.mode
        b_size = self.b_size
        
        if (n_iter == 0):
            n_iter = self.n_iter
        else:
            self.n_iter = n_iter
        
        if (step_size == 0):
            step_size = self.step_size
        else:
            self.step_size = step_size
        
        if (reg == 0):
            reg = self.reg
        else:
            self.reg = reg
        
        #Set random seed for repeatability
        np.random.seed(self.seed)
        
        
        #Define some local varaibles
        D = X_trg.shape[1] #Number of features
        K = np.unique(y_trg).size #Number of classes assuming class index starts from 0

        # Start with an initial set of parameters randomly
        self.W1 = 0.01 * np.random.randn(D,h)
        self.b1 = np.zeros((1,h))

        self.W2 = 0.01 * np.random.randn(D,h)
        self.b2 = np.zeros((1,h))

        self.Wh = 0.01 * np.random.randn(h,K)
        self.bh = np.zeros((1,K))

        # Set batch size
        batch_size = int(X_trg.shape[0]/b_size)
        max_row = X_trg.shape[0]  #Tot rows in training data
        batch_count=0
        for row in range(0, max_row, batch_size):
            #Compute batch to process
            tmp = row + batch_size
            if tmp > max_row:
                batch_max = max_row
                batch_size = batch_max - row
            else:
                batch_max = tmp

            # Get the batch of data points
            X_batch = X_trg[row:batch_max]
            y_batch = y_trg[row:batch_max]

            batch_count += 1
            if mode.upper()=='V': 
                print("Batch Count:", batch_count)

            # gradient descent loop
            for i in range(n_iter):

                # evaluate class scores, [N x K]
                tmp1 = np.dot(X_batch, self.W1) + self.b1
                tmp2 = np.dot(X_batch, self.W2) + self.b2
                hidden_layer_out = np.maximum(tmp1, tmp2) # note, maxout activation
                scores = np.dot(hidden_layer_out, self.Wh) + self.bh

                # compute the class probabilities
                exp_scores = np.exp(scores)
                probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]

                # compute the loss: average cross-entropy loss and regularization
                corect_logprobs = -np.log(probs[range(batch_size),y_batch])
                data_loss = np.sum(corect_logprobs)/batch_size
                reg_loss = 0.5*reg*np.sum(self.W1*self.W1) \
                + 0.5*reg*np.sum(self.W2*self.W2) \
                + 0.5*reg*np.sum(self.Wh*self.Wh)
                loss = data_loss + reg_loss
                if (mode.upper()=='V') and ((i % (n_iter/10) == 0) or (i % (n_iter-1) == 0)):
                    print ("iteration {}: loss {}".format(i, loss))

                # compute the gradient on scores
                dscores = probs
                dscores[range(batch_size),y_batch] -= 1
                dscores /= batch_size

                # backpropate the gradient to the parameters
                # first backprop into parameters Wh and bh
                dWh = np.dot(hidden_layer_out.T, dscores)
                dbh = np.sum(dscores, axis=0, keepdims=True)

                # next backprop into hidden layer
                dhidden = np.dot(dscores, self.Wh.T)

                # backprop the maxout non-linearity

                #derivative of hidden layer appears differently to the two input layers
                dhid1 = dhidden
                tmp1 = np.dot(X_batch, self.W1) + self.b1
                dhid1[hidden_layer_out > tmp1] = 0
                # Now into W1,b1
                dW1 = np.dot(X_batch.T, dhid1)
                db1 = np.sum(dhid1, axis=0, keepdims=True)

                dhid2 = dhidden
                tmp2 = np.dot(X_batch, self.W2) + self.b2
                dhid2[hidden_layer_out > tmp2] = 0
                # finally into W2,b2
                dW2 = np.dot(X_batch.T, dhid2)
                db2 = np.sum(dhid2, axis=0, keepdims=True)

                # add regularization gradient contribution
                dWh += reg * self.Wh
                dW1 += reg * self.W1
                dW2 += reg * self.W2

                # perform a parameter update
                self.W1 += -step_size * dW1
                self.b1 += -step_size * db1
                self.W2 += -step_size * dW2
                self.b2 += -step_size * db2
                self.Wh += -step_size * dWh
                self.bh += -step_size * dbh

    def score(self, X_tst, y_tst):
        # Run on test data 
        tmp1 = np.dot(X_tst, self.W1) + self.b1
        tmp2 = np.dot(X_tst, self.W2) + self.b2
        hl_out = np.maximum(tmp1, tmp2) # maxout activation
        scores = np.dot(hl_out, self.Wh) + self.bh

        predicted_class = np.argmax(scores, axis=1)
        acc = (np.mean(predicted_class == y_tst))
    
        return acc
```
